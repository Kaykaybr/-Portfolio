```python
# Rename columns 
df.rename(columns={'Q1 - Which Title Best Fits your Current Role?': 'Current_Role', 
                   'Q2 - Did you switch careers into Data?': 'Switch_Careers',
                   'Q3 - Current Yearly Salary (in USD)': 'Current_Yearly_Salary(USD)',
                   'Q4 - What Industry do you work in?': 'Industry',
                   'Q5 - Favorite Programming Language': 'Fav_Programming_Lang',
                   'Q6 - How Happy are you in your Current Position with the following? (Salary)': 'Satisfication(salary)',
                   'Q6 - How Happy are you in your Current Position with the following? (Work/Life Balance)': 'Satisfication(Work/Life_Balance)',
                   'Q6 - How Happy are you in your Current Position with the following? (Coworkers)': 'Satisfication(Coworkers)',
                   'Q6 - How Happy are you in your Current Position with the following? (Management)': 'Satisfaction(Management)',
                   'Q6 - How Happy are you in your Current Position with the following? (Upward MoBIlity)': 'Satisfaction(UpwardMobility)',
                   'Q6 - How Happy are you in your Current Position with the following? (Learning New Things)': 'Satisfaction(LearningNewThings)',
                   'Q7 - How difficult was it for you to break into Data?': 'Difficulty_breaking_into_Data',
                   'Q8 - If you were to look for a new job today, what would be the most important thing to you?': 'Important_when_looking_fora_new_job',
                   'Q9 - Male/Female?': 'Gender',
                   'Q10 - Current Age': 'Age',
                   'Q12 - Highest Level of Education': 'Highest_Level_of_Education',
                   'Q13 - Ethnicity': 'Ethnicity',
                   'Q11 - Which Country do you live in?': 'Country_of_Residence'
                  }, inplace=True)

df
```

```python
df.isna().sum()
```

```python
df['Highest_Level_of_Education'] = df['Highest_Level_of_Education'].fillna('Bachelors')

df['Industry'] = df['Industry'].fillna('IT')

df['Fav_Programming_Lang'] = df['Fav_Programming_Lang'].fillna('None')

df['Satisfication(salary)'] = df['Satisfication(salary)'].fillna(1)

df['Satisfication(Work/Life_Balance)'] = df['Satisfication(Work/Life_Balance)'].fillna(1)

df['Satisfication(Coworkers)'] = df['Satisfication(Coworkers)'].fillna(1)

df['Satisfaction(Management)'] = df['Satisfaction(Management)'].fillna(1)

df['Satisfaction(UpwardMobility)'] = df['Satisfaction(UpwardMobility)'].fillna(1)

df['Satisfaction(LearningNewThings)'] = df['Satisfaction(LearningNewThings)'].fillna(1)

df['Important_when_looking_fora_new_job'] = df['Important_when_looking_fora_new_job'].fillna('None')
```

```python
df
```

```python
df.isna().sum()
```

```python
# splitting the salary column
df2 = df['Current_Yearly_Salary(USD)'].str.split('-', expand=True)

# Rename columns
df2.columns = ['lower_limit_salary', 'upper_limit_salary'] 

# remove the k and +
df2['lower_limit_salary'] = df2['lower_limit_salary'].str.replace('k', '', regex=False)
df2['lower_limit_salary'] = df2['lower_limit_salary'].str.replace('+', '', regex=False)
df2['upper_limit_salary'] = df2['upper_limit_salary'].str.replace('k', '', regex=False)
                                          
# convert datatype
df2['lower_limit_salary'] = df2['lower_limit_salary'].astype('int64')    

# filling in empty values
df2['upper_limit_salary'] = df2['upper_limit_salary'].fillna(0)

# convert datatype
df2['upper_limit_salary'] = df2['upper_limit_salary'].astype('int64')

# changing some values
df2.loc[df2['upper_limit_salary'] == 0] = 225

# creating an average salary column
df2['average_salary'] = (df2['lower_limit_salary'] + df2['upper_limit_salary'])/2

df2
```

```python
df = df.join(df2)

df.head()
```

```python
# format date
df['Date Taken (America/New_York)'] = pd.to_datetime(df['Date Taken (America/New_York)'], errors='coerce')
```

```python
df_numerized = pd.DataFrame(df)

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category').cat.codes

df_numerized.head()
```

```python
corr_matrix = df_numerized.corr()

sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
```

```python
corr_pairs = corr_matrix.unstack()

sorted_pairs = corr_pairs.sort_values()

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
