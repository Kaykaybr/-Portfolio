```python
df = movie_data

df.head()
```

```python
df['votes'] = df['votes'].astype('int64')

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head()
```

```python
# Splitting the released column

# Creating a dataframe for released
split_col = df['released']
df2 = pd.DataFrame(split_col)

# Split the column
df_split = df2['released'].str.split('(', expand=True)

# Rename the new columns
df_split.columns = ['released_date', 'country_of_release']

# Remove closed brackets
df_split['country_of_release'] = df_split['country_of_release'].str.replace(')', '', regex=False)


# Combine with the original DataFrame
df = df.join(df_split)

df.head()
```

```python
# Standardize the release date 

# Remove trailing spaces 
df['released_date'] = df['released_date'].str.strip() 

# Convert to datetime 
df['released_date'] = pd.to_datetime(df['released_date'], errors='coerce') 

# Format the date 
df['released_date'] = df['released_date'].dt.strftime('%d-%m-%Y')

df.head()
```

```python
# Create correct year column
df['release_year']=df['released_date'].astype(str).str[6:]

df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Correlation between budget and gross

sns.regplot(x=df['budget'], y=df['gross'], data=df, scatter_kws={"color":"blue"})
plt.title('Budget VS Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
```

```python
# Looking at correlation

numeric_cols = df.select_dtypes(include=[float, int])

numeric_cols.corr()
```

```python
corr_matrix = numeric_cols.corr()

sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
```

```python
# Cat coding object type columns

df_numerized = pd.DataFrame(df)

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category').cat.codes

# Cat coding Name specifically
if (df_numerized['name'].dtype == 'category'):
    df_numerized['name'] = df_numerized['name'].cat.codes

df_numerized.head()
```

```python
# numeric_cols = df_numerized.select_dtypes(include=[float, int])

df_numerized.corr()

corr_matrix = df_numerized.corr()

sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
```

```python
correlation = df_numerized.corr()

corr_pairs = correlation.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
